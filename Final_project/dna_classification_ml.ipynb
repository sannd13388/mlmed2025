# DNA CLASSIFICATION
In this project ,we will be exploring what of bioinformatics.

For this project, we wil use a dataset from the UCI Machine Learning response for a story of that has 100 DNA sequences with 57 sequential nucleotides.

We will learn how to import data from the UCI repositories that can put to numerical data below and train classification algorithms and compare andcontrast classification machine learning algorithms.

```python
import sys
import numpy as np
import sklearn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/molecular-biology/promoter-gene-sequences/promoters.data"
```

```python
names = ["Class", "id", "Sequence"]
```

```python
data = pd.read_csv(url, names = names)

data
```

```python
classes = data.loc[:,"Class"]

print(classes)
```

```python
sequences = list(data.loc[: , "Sequence"])
dataset = {}

for i, seq in enumerate(sequences):

  nucleotides = list(seq)
  nucleotides = [x for x in nucleotides if x != '\t']

  nucleotides.append(classes[i])

  dataset[i] = nucleotides

print(dataset[0])
```

```python
dframe = pd.DataFrame(dataset)

dframe
```

```python
df = dframe.transpose()

df.iloc[:5]
```

```python
df.rename(columns = {57 : "Class"}, inplace = True)
df
```

```python
numerical_df = pd.get_dummies(df, dtype = int)
numerical_df.head()
```

```python
df = numerical_df.drop(columns=['Class_-'])

df.rename(columns = {'Class_+': 'Class'}, inplace = True)
df.head()
```

```python
from sklearn import model_selection

X = np.array(df.drop(['Class'], axis=1))
y = np.array(df['Class'])

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

scoring ='accuracy'

names = ["Nearest Neighbors", "Gaussian Process","Decision Tree","Random Forest",
         "Neural Net", "AdaBoost","Naive Bayes","SVM Linear","SVM RBF","SVM Sigmoid", "Logistic Regression"]
classifiers =[
    KNeighborsClassifier(n_neighbors=3),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10,max_features=1),
    MLPClassifier(alpha=1),
    AdaBoostClassifier(),
    GaussianNB(),
    SVC(kernel='linear'),
    SVC(kernel='rbf'),
    SVC(kernel='sigmoid'),
    LogisticRegression()
]
models =zip(names,classifiers)

results = []
names=[]
accuracy = []
for name,model in models:
    kfold = model_selection.KFold(n_splits = 10)
    cv_results = model_selection.cross_val_score(model, X_train,y_train,cv=kfold,scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg= "%s: %f (%f)" %(name, cv_results.mean(), cv_results.std())
    
    print(msg)
    model.fit(X_train,y_train)
    predictions= model.predict(X_test)
    print(name)
    print( classification_report(y_test,predictions))
```

```python
fig = plt.figure(figsize=(10,12), dpi = 80)
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results, vert=False)
ax.set_yticklabels(names)
plt.show()
plt.savefig("algorithm_comparison.png")
```
